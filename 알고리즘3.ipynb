### 2번 (headq 사용) - 구현 못함

```python
def solution(jobs):
    import heapq
    answer = 0
    time = 0
    n = 0
    heapq.heapify(jobs)
    while jobs:
        req, proc = heapq.heappop(jobs)

        answer = proc if n == 0 else answer + time - req + proc
        time = req + proc if n ==0 else time + proc
        n += 1

        heap = []
        while jobs and jobs[0][0] < time:
             r, p = heapq.heappop(jobs) 
             heapq.heappush(heap, (-p, r))

        while heap:
            p, r = heapq.heappop(heap)
            jobs.insert(0, [r, -p])

    answer //= n
    return answer
```

```python
def solution(arguments):
    queue = list()
    for operation in arguments :
        if operation[0] == 'I' :
            queue.append(int(operation[1 : ]))
        elif operation[0] == 'D' :

            if len(queue) == 0 :
                continue

            if int(operation[1 : ]) == 1 :
                queue.remove(max(queue))
            elif int(operation[1 : ]) == -1 :
                queue.remove(min(queue))

    if len(queue) == 0 :
        answer = [0, 0]
    else :
        answer = [max(queue), min(queue)]

    return answer
```
